```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ironman_lake_placid_female_2022_canadian.csv')
df.head()
```

```python
dff = df.loc[df['Division'] == 'F25-29', ['Name','Swim.Time', 'Bike.Time', 'Run.Time']]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = dff.Name
y1 = dff['Swim.Time']
y2 = dff['Bike.Time']
y3 = dff['Run.Time']

w = 0.8
# Plotting
plt.bar(x, y1, w, color='r')
plt.bar(x, y2, w, bottom=y1, color='b')
plt.bar(x, y3, w, bottom=y1+y2, color='y')


# Labels and Title
plt.xlabel("Corredoras Triatlon 25-29 años")
plt.ylabel("Tiempo en minutos")
plt.legend(["Natación", "Bicicleta", "Correr"])
plt.title("Tiempos Corredoras Triatlón")
plt.xticks(rotation=60)
# Display
plt.show()
```

## Las tecnologías de energía limpia requieren una amplia gama de minerales y metales
Las tecnologías de energía limpia, desde turbinas eólicas y paneles solares hasta vehículos eléctricos y almacenamiento de baterías, requieren una amplia gama de minerales y metales. El tipo y el volumen de las necesidades minerales varían ampliamente en todo el espectro de tecnologías de energía limpia, e incluso dentro de una determinada tecnología (por ejemplo, la química de las baterías de los vehículos eléctricos).

https://www.iea.org/reports/the-role-of-critical-minerals-in-clean-energy-transitions/mineral-requirements-for-clean-energy-transitions

```python
# Necesidades minerales críticas para tecnologías de energía limpia con alta importancia
Solar_fotovoltaica = set(['Cobre', 'Aluminio', ''])
Eolica = set(['Cobre', 'Tierras raras', 'Zinc'])
baterias_EV = set(['Cobre', 'Cobalto', 'Niquel', 'Litio', 'Tierras raras', 'Aluminio'])

from matplotlib_venn import venn3
venn3([Solar_fotovoltaica, Eolica, baterias_EV], ('Solar_fotovoltaica', 'Eolica', 'baterias_EV'))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(0, 10, 0.1)  # x-axis values
y1 = np.sin(x)  # series 1
y2 = np.cos(x)  # series 2
y3 = np.random.normal(size=len(x))  # series 3

# Create the figure and axis
fig, ax = plt.subplots()

# Define the colors and alpha values for each series
colors = ['C0', 'C1', 'C2']  # color cycle
alphas = [0.5, 0.7, 0.9]  # transparency

# Plot each series as a filled area
for i, (y, color, alpha) in enumerate(zip([y1, y2, y3], colors, alphas)):
    ax.fill_between(x, 0, y, color=color, alpha=alpha)

# Set axis labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Value')
ax.set_title('Horizon Graph')

# Show the plot
plt.show()
```
